# Population Distribution Visualization
### Data Source: World Bank (SP.POP.TOTL)


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import os
os.listdir()
```

```python
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_174326.csv", skiprows=4)
```

```python
print(df.head())
```

```python
population_2022 = df[['Country Name', '2022']].dropna()
```

```python
sample_countries = population_2022.head(10)
```

```python
plt.figure(figsize=(10,6))
plt.bar(sample_countries['Country Name'], sample_countries['2022'])
plt.xticks(rotation=45)
plt.title("Population Distribution of Sample Countries (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.show()

```

```python
plt.hist(population_2022['2022'], bins=20)
plt.title("Population Distribution Histogram (2022)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.show()
```

Conclusion

This project successfully visualizes population distribution using a bar chart. The World Bank dataset provides reliable population data, and Python libraries allow easy and effective visualization.
